```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
driver = webdriver.Chrome()
driver.get("https://nypost.com/environment/")
#while true:
try:
    close_popups = driver.find_element(By.ID, "zephrModalClose")
    close_popups.click()
    time.sleep(3)
    print("pop up closed!")
except Exception as e:
    print("couldn't close pop-up:", e)
    
while True:
    try:
        button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "button.button.button--solid.see-more"))
        )
        button.click()
        print("Loaded more articles!")
        time.sleep(2)
    except (TimeoutException, NoSuchElementException):
        print("No more articles to load  finished scrolling.")
        break
    except ElementClickInterceptedException:
        print("Something blocked the button click  retrying.")
        time.sleep(2)
        continue
    
html_content = driver.page_source
driver.quit()
```

```python
html_content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')
page_section = soup.find('div', class_ = 'section__content')
stories = page_section.find_all('div', class_ = ['story__text', 'story__inner'])
headlines = []
excerpts = []
dates = []

for i in stories:
    headline = i.find(['h3','h2'])
    if headline is not None:
        headlines.append(headline.get_text(strip=True))
    else: 
        headlines.append('n/a')

for i in stories:
    excerpt = i.find('p')
    if excerpt is not None:
        excerpts.append(excerpt.get_text(strip=True))
    else: 
        excerpts.append('n/a')
        
for i in stories:
    date = i.find('span')
    if date is not None:
        dates.append(date.get_text(strip=True))
    else: 
        dates.append('n/a')
```

```python
del headlines[1]
del excerpts[1]
del dates[1]
```
